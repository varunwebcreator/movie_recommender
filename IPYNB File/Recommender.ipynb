```python
# function to extract Genre out of the movie which is stored inside name key of genre column

def convert(text):
    # empty list to store the data 
    L = []

    # loop to extract genre
    for i in ast.literal_eval(text):
        L.append(i['name']) 
        
    return L 
```

```python
# replacing the genre column value with actual genre value rather than a dictionary
movies['genres'] = movies['genres'].apply(convert)
movies.head()
```

```python
movies['keywords'] = movies['keywords'].apply(convert)
movies.head()
```

```python
# function to separate top 3 cast name of the movie

def convert3(text):
    L = []
    counter = 0
    
    for i in ast.literal_eval(text):
        if counter < 3:
            L.append(i['name'])
        counter+=1
    
    return L 
```

```python
movies['cast'] = movies['cast'].apply(convert)
movies.head()
```

```python
# selecting the top 3
movies['cast'] = movies['cast'].apply(lambda x:x[0:3])
```

```python
movies.head()
```

```python
# function to extract the director_name 

def fetch_director(text):
    L = []
    
    for i in ast.literal_eval(text):
        if i['job'] == 'Director':
            L.append(i['name'])
    
    return L 
```

```python
movies['crew'] = movies['crew'].apply(fetch_director)
movies.head()
```

```python
# converting overview column from str to list
movies['overview'] = movies['overview'].apply(lambda x:x.split())
movies.sample(5)
```

```python
# fynction to remove extra space in any column
def collapse(L):
    L1 = []
    
    for i in L:
        L1.append(i.replace(" ",""))
    
    return L1
```

```python
# applying the collapse function on these 4 columns
movies['cast'] = movies['cast'].apply(collapse)
movies['crew'] = movies['crew'].apply(collapse)
movies['genres'] = movies['genres'].apply(collapse)
movies['keywords'] = movies['keywords'].apply(collapse)
```

```python
movies.head()
```

```python
# new column in data all the mentioned columns data
movies['tags'] = movies['overview'] + movies['genres'] + movies['keywords'] + movies['cast'] + movies['crew']
```

## Text_Vectorization

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
# converting text into vectors
cv = CountVectorizer(max_features = 5000,stop_words = 'english')
```

```python
# vector transformation
vectors = cv.fit_transform(new_df['tags']).toarray()
```

```python
# vector shape
vectors.shape
```

## Stemming

```python
import nltk
from nltk.stem.porter import PorterStemmer
```

```python
ps = PorterStemmer()
```

```python
# stemming function
def stem(text):
    y = []

    for word in text.split():
        y.append(ps.stem(word))

    return " ".join(y)
```

```python
# applying the function 
new_df['tags'] = new_df['tags'].apply(stem)
```

```python
movies.head()
```

## Recommender

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
# cosine_similarity calculates the cosine of the angle between two vectors in a multi-dimensional space. 
# It is a measure of similarity between two vectors, often used in text and document analysis, recommendation systems, and clustering tasks.
```

```python
# similarity (using cosine_similarity as eculdian distance is not good option for high dimension data)
# cosine similarity will calcualte the similarity on the bases of distance
similarity = cosine_similarity(vectors)
```

```python
# recommender function for movie recommendation
def recommend(movie):
    
    index = new_df[new_df['title'] == movie].index[0]
    
    distances = sorted(list(enumerate(similarity[index])),reverse=True,key = lambda x: x[1])
    
    for i in distances[1:6]:
        print(new_df.iloc[i[0]].title)
```

```python
recommend('Gandhi')
```
